```python
import sys

if 'scalar_fastapi' not in sys.modules:
    !pip install -q fastapi uvicorn scalar-fastapi

from fastapi import FastAPI
import uvicorn
from fastapi.responses import HTMLResponse
from scalar_fastapi.scalar_fastapi import get_scalar_api_reference
import threading
import time

app = FastAPI(
    title="My Simple FastAPI App",
    description="A basic FastAPI server with Scalar API documentation.",
    version="1.0.0",
    docs_url=None, # Disable default Swagger UI
    redoc_url=None # Disable default ReDoc
)

# Integrate Scalar for API documentation by creating a new route.
# This serves Scalar docs at the root path.
@app.get("/", response_class=HTMLResponse, include_in_schema=False)
async def scalar_api_docs():
    return get_scalar_api_reference(
        title="My Simple FastAPI App Docs",
        description="Interactive API documentation powered by Scalar.",
        version="1.0.0",
        openapi_url="/openapi.json" # Point to FastAPI's openapi spec
    )

@app.get("/")
async def read_hello():
    return {"message": "Hello from FastAPI!", "docs_root": "You are seeing Scalar docs at the root of this server. Try /items/{item_id}."}

@app.get("/items/{item_id}")
async def read_item(item_id: int, q: str = None):
    return {"item_id": item_id, "q": q}

# Function to run the Uvicorn server in a separate thread
def run_uvicorn():
    # Use log_level="error" or "warning" to reduce verbose output in Colab
    uvicorn.run(app, host="0.0.0.0", port=8265, log_level="warning")

# Start Uvicorn in a new thread so the cell doesn't block the Colab kernel.
# The `daemon = True` ensures the thread will close if the main program exits.
print("Starting FastAPI server on port 8265...")
uvicorn_thread = threading.Thread(target=run_uvicorn)
uvicorn_thread.daemon = True
uvicorn_thread.start()

# Give some time for the server to start up
time.sleep(2)
print("FastAPI server should now be running.")
print("The Scalar API documentation is accessible in the iframe above (which points to http://localhost:8265/).")
print("You can also access the '/hello' endpoint or '/items/{item_id}' endpoints directly.")
```
